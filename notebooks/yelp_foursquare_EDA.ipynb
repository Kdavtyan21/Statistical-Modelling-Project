```python
# imports
import requests
import pandas as pd
import os
```

# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
df = pd.read_excel('dataframe.xlsx')
data_fsq = []
i = 0
api_key_fsq = os.environ["FOURSQUARE_API_KEY"]
latitude = df["Latitude"][i]
longitude = df["Longitude"][i]
url = "https://api.foursquare.com/v3/places/search?ll=" + str(latitude) + "%2C" + str(longitude) + "&radius=1000"

headers = {
    "accept": "application/json",
    "Authorization": api_key_fsq
}

r = requests.get(url, headers=headers)
data_fsq = r.json()
    
```

```python
data_fsq[1]
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
location_data = []
for poi in data_fsq["results"]:
    location_info = {
        "name": poi["name"],
        "address":poi["location"]["address"],
        "latitude": poi["geocodes"]["main"]["latitude"],
        "longitude": poi["geocodes"]["main"]["longitude"]
    } 
    location_data.append(location_info)
location_data
```

Put your parsed results into a DataFrame

```python
df_four = pd.DataFrame(location_data)

# Display the DataFrame
df_four
```

# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
url = "https://api.yelp.com/v3/businesses/search?latitude=" + str(latitude) + "&longitude=" + str(longitude) + "&radius=1000"
api_key_yelp = os.environ["YELP_API_KEY"]
headers = {
    "accept": "application/json",
    "Authorization": "Bearer " + api_key_yelp
}

response = requests.get(url, headers=headers)
data_yelp = response.json()
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
location_data_yelp = []
for poi in data_yelp["businesses"]:
    location_info_yelp = {
        "name": poi["name"],
        "rating": poi["rating"],
        "address":poi["location"]["address1"],
        "latitude": poi["coordinates"]["latitude"],
        "longitude": poi["coordinates"]["longitude"]
    } 
    location_data_yelp.append(location_info_yelp)
location_data_yelp
```

```python
df_yelp = pd.DataFrame(location_data_yelp)
df_yelp
```
